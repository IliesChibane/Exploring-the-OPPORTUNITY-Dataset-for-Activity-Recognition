```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('preprocessed_data.csv')

df
```

```python
echantillon = []
classes = df["HL_Activity"].unique()
for c in classes:
    sequence = []
    hl = list(df.loc[df['HL_Activity'] == c].index)
    for i in hl:
        if i+1 in hl:
            sequence.append(i)
        else:
            break
    echantillon.append(sequence)
for e in echantillon:
    print(len(e))
```

```python
fig, axes = plt.subplots(2, 3,figsize=(15, 7))
i = 0
j = 0
l = 0
random_column = df[df.columns[23]].to_numpy()
classes = ["Relaxing", "Coffee time", "Early morning", "Cleanup", "Sandwich time"]
for e in echantillon:
    axes[i, j].plot(random_column[e[0]:e[-1]])
    axes[i, j].set_title(classes[l])
    l += 1
    j += 1
    if j == 3:
        j = 0
        i += 1
fig.suptitle(df.columns[23])
plt.show()
```
